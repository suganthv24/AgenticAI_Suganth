```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('salary_dataset.csv')
df.drop(columns=['Unnamed: 0'], inplace=True)
df.info()
df.describe()
df.isnull().sum()
df.duplicated().sum()
```

```python
x = df['YearsExperience'].values.reshape(-1, 1)
y = df['Salary']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LinearRegression()

model.fit(x_train, y_train)

predicted_Salary = model.predict(x_test)

```

```python
#finding the accuracy of the model
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error,mean_absolute_percentage_error
accuracy = r2_score(y_test, predicted_Salary)
print(f'R2 Score: {accuracy}')

mse = mean_squared_error(y_test, predicted_Salary)
print(f'Mean Squared Error: {mse}')

mae = mean_absolute_error(y_test, predicted_Salary)
print(f'Mean Absolute Error: {mae}')

mape = mean_absolute_percentage_error(y_test, predicted_Salary)
print(f'Mean Absolute Percentage Error: {mape}')
```

```python
plt.scatter(x_train, y_train, color='blue')
plt.plot(x_test, predicted_Salary, color='red')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')    
plt.title('Years of Experience vs Salary')
plt.show()
```
