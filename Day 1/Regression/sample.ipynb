### 1. Regression
Regression models predict a continuous output based on input features. They are used when the target variable is a real number. Common algorithms include Linear Regression, Polynomial Regression, and Support Vector Regression.

Steps to build a regression model:
1. Gather all the required data.


2. Define the features(X) and the target(Y).
3. Splitting the data into training and testing sets.
4. Train the model using the training data.
5. Evaluate the model using the testing data.
6. Visualize the results.

## Step 1: Importing the required Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

### Experimenting with random data

```python
plt.style.use("ggplot")
```

```python
random_data = np.random.RandomState(1)
x = 10 * random_data.rand(50)
y = 2 * x - 5 + random_data.rand(50)
plt.scatter(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Random Data")
plt.show()
```

```python
from turtle import color


model = LinearRegression()
model.fit(x[:, np.newaxis],y)

xfit = np.linspace(0,10,100)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x,y,color="blue")
plt.plot(xfit,yfit,color="black")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression Fit")
plt.show()
```
