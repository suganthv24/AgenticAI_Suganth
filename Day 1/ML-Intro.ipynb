<h1 style="color:#FFD740;">
Experimenting with random data
</h1>

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use("ggplot")
import numpy as np
```

```python
randomdata  = np.random.RandomState(1)
x= 10 * randomdata.rand(50)
y= 2 * x -5 + randomdata.rand(50)
plt.scatter(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Random Data")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x[:, np.newaxis],y) # fit the model to the data

xfit = np.linspace(0, 10 ,1000) # create a sequence of 1000 numbers between 0 and 10
yfit = model.predict(xfit [:, np.newaxis]) # predict the y values for the x values
plt.scatter(x,y, label="Data") # plot the original data

plt.plot(xfit, yfit, color="green", label="Line of Best Fit") #plot the line of best fit
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Random Data with Line of Best Fit")
plt.legend()
plt.show()


```
