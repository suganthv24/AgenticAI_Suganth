```python
import csv
import random

f = "movie-dialog-corpus/movie_lines.tsv"

texts = []
with open(f, "r", encoding="utf-8", errors="ignore") as i:
    reader = csv.reader(i, delimiter="\t")
    for row in reader:
        texts.append(row[-1])

random.seed(1)
random.shuffle(texts)
texts = texts[:10000]
texts[:10]
```

```python
import tensorflow_hub as hub
import tensorflow as tf
import numpy as np
```

```python
tf.compat.v1.logging.set_verbosity(tf.compat.v1.logging.ERROR)
embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")
```

```python
def embed_texts(texts):
    
    tf.compat.v1.logging.set_verbosity(tf.compat.v1.logging.ERROR)

    with tf.compat.v1.Session() as session:
        session.run([tf.compat.v1.global_variables_initializer(), tf.compat.v1.tables_initializer()])
        embeddings = session.run(embed(texts))

    return np.array(embeddings).tolist()

embeddings = embed_texts(texts)
```

```python


module_url = "https://tfhub.dev/google/universal-sentence-encoder/4"
model = hub.load(module_url)
print("Module loaded.")
def embed(input):
    return model(input)
```

```python
message = ["word", "sentence", "paragraph"]

message_embeddings = embed(message)

for i, message_embedding in enumerate(np.array(message_embeddings).tolist()):
    print("Message: {}".format(message[i]))
    print("Embedding size: {}".format(len(message_embedding)))
    message_embedding_snippet = ", ".join(
        [str(x) for x in message_embedding[:3]]
    )
    print("Embedding (first 3 values): {}".format(message_embedding_snippet))
```
